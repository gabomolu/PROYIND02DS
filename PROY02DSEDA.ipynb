```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Rutas de entrada y salida
input_dir = r"C:\Users\GabrielArturoMonzónL\Desktop\PROYECTO2\DATASET\CSV"
output_dir = r"C:\Users\GabrielArturoMonzónL\Desktop\PROYECTO2\DATASET\CSV_EDA"

# Crear la carpeta de salida si no existe
os.makedirs(output_dir, exist_ok=True)

# Función para realizar el EDA en un archivo
def realizar_eda(filename):
    # Definir rutas de archivo
    input_path = os.path.join(input_dir, filename)
    output_path = os.path.join(output_dir, f"{os.path.splitext(filename)[0]}_eda.csv")

    # Cargar el dataset
    df = pd.read_csv(input_path)

    # Vista general del dataset
    print(f"\nProcesando archivo: {filename}")
    print("Primeras filas del dataset:")
    print(df.head())
    print("\nInformación general:")
    print(df.info())
    print("\nEstadísticas descriptivas:")
    print(df.describe())

    # 1. Identificar valores nulos
    print("\nValores nulos por columna:")
    print(df.isnull().sum())

    # Imputar valores nulos (ejemplo: reemplazar con la mediana para columnas numéricas)
    df.fillna(df.select_dtypes(include=[np.number]).median(), inplace=True)

    # 2. Verificar tipos de datos y convertir si es necesario
    # Aquí puedes especificar conversiones, si alguna columna tiene un tipo incorrecto
    # Ejemplo: df['columna_fecha'] = pd.to_datetime(df['columna_fecha'], errors='coerce')

    # 3. Análisis de correlación para columnas numéricas
    numeric_df = df.select_dtypes(include=[np.number])  # Solo selecciona columnas numéricas
    correlation_matrix = numeric_df.corr()
    print("\nMatriz de correlación:")
    print(correlation_matrix)

    # Visualizar la matriz de correlación
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f"Matriz de correlación - {filename}")
    plt.show()

    # 4. Distribuciones de variables numéricas
    numeric_df.hist(bins=15, figsize=(15, 10))
    plt.suptitle(f"Distribuciones de variables numéricas - {filename}")
    plt.show()

    # 5. Guardar dataset después de EDA
    df.to_csv(output_path, index=False)
    print(f"Dataset procesado guardado en: {output_path}")

# Procesar todos los archivos en el directorio de entrada
for file in os.listdir(input_dir):
    if file.endswith(".csv"):
        realizar_eda(file)
```
